Note: The random_state=2 ensures that the random splitting of the data into training and testing sets will be the same if you run the code multiple times with the same dataset. If you change the random_state value, you may get a different random split.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df.iloc[:,0],test_size=0.2, random_state=2)
```

```python
X_train.shape
```

Converting everything to the same scale

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)             #we didn't use fit_transform because we don't want to recalculate the mean and sd
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5)       #creating an object
```

```python
knn.fit(X_train, y_train)       #trained our data
```

```python
from sklearn.metrics import accuracy_score

y_pred = knn.predict(X_test)     #predict

accuracy_score(y_test, y_pred)   #simultaneously check the accuracy score
```
